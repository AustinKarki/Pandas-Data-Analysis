```python
def generate_rating_df(df):
  rating_df=df.groupby(["cp","slope"]).agg({"id":"count"}).reset_index()
  rating_df=rating_df[rating_df["id"]!=0]
  rating_df.columns=["cp","slope","counts"]
  rating_df=rating_df.sort_values("slope")
  # now we will try to calculate the percentages
  total=rating_df['counts'].sum()
  rating_df['percentage']=rating_df['counts']/total*100
  return rating_df

rating_df=generate_rating_df(df)
fig=px.bar( rating_df,x="cp",y="counts",color="slope",barmode="stack",text="percentage")
fig.update_traces(textposition="outside",textfont_size=16,texttemplate="%{text:.2f}")
fig.update_layout(barmode="group",xaxis_title="CP",legend_title="Slope",yaxis_title="Counts")
fig.update_layout(height=550,width=1000,title_text="Distribution of Chest Pain Type by Percentages",title_font_size=24)
fig.show()
```
